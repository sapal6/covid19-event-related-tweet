```
# Import & Settings
import os

!pip install pyDataverse 
#documentation here: https://pydataverse.readthedocs.io/en/latest/index.html#quickstart
from pyDataverse.api import Api
from pyDataverse.models import Dataverse
```

```
# Establish connection to Dataverse
base_url = 'https://dataverse.harvard.edu'
api = Api(base_url)
print('Dataverse connection status: ', api.status)
```

```
# Get dataverse info
dv = 'data-cradle' 
response = api.get_dataverse(dv)
response.json()
```

```
%time
# Get dataverse information by id
dv = 'doi:10.7910/DVN/CE2530'
response = api.get_dataset(dv) # in Dataverse 
dataset_id = response.json()['data']['id']
print('Dataset ID: ', dataset_id, '\n')

datafile_ids = []
labels = []
for file_index in range(len(response.json()['data']['latestVersion']['files'])):
    labels.append(response.json()['data']['latestVersion']['files'][file_index]['label'])
    datafile_ids.append(response.json()['data']['latestVersion']['files'][file_index]['dataFile']['id'])
files = dict(zip(labels,datafile_ids))
print('Datafile ID: ', files, '\n')

print('Response results: ')
response.json()
```

```
%time
for label in files:
    response = api.get_datafile(files[label])
    
    #Assuming that a length less than 4k results in an error --> return that error response
    if len(response.content) < 4000:
        print(response.content)
    
    with open(label, 'wb') as f: 
        f.write(response.content)
```
